# Plots for the Appendix B: Mixed-type PDE

```python
from rcparams import *
import numpy as np
```

```python
def tricomi(s,c,xvals):
    return np.array([c + s*(2.0/3.0)*pow(-x,1.5) for x in xvals])

def keldysh(s,c,xvals):
    return np.array([c + s*(2.0)*pow(-x,0.5) for x in xvals])
```

```python
x = np.arange(-5,0,0.001)

for c in np.arange(-30,31,8):
    plt.plot(x,tricomi( 1,c,x),'k--')
    plt.plot(x,tricomi(-1,c,x),'m-')

plt.fill([0,0,10,10],[-30,30,30,-30],'b',alpha=0.1)
plt.vlines(0,-20,20,'k')
    
plt.title(r'Tricomi: $\partial_x^2u + x\partial_y^2u=0$')
plt.xlim(-5,2)
plt.ylim(-20,20)
plt.xlabel(r'x')
plt.ylabel(r'y')

plt.savefig('tricomi_integral_curves.pdf', format='pdf', bbox_inches='tight')
```

```python
x = np.arange(-5,0,0.0001)

for c in np.arange(-30,31,8):
    plt.plot(x,keldysh( 1,c,x),'k--')
    plt.plot(x,keldysh(-1,c,x),'m-')

plt.fill([0,0,10,10],[-30,30,30,-30],'b',alpha=0.1)
plt.vlines(0,-20,20,'k')
    
plt.title(r'Keldysh: $x\partial_x^2u + \partial_y^2u=0$')
plt.xlim(-5,2)
plt.ylim(-20,20)
plt.xlabel(r'x')
plt.ylabel(r'y')

plt.savefig('keldysh_integral_curves.pdf', format='pdf', bbox_inches='tight')
```
